```python

import  speech_recognition as sr
import sounddevice as sd
import numpy as np
import time
import win32com.client
import wikipedia
import webbrowser

import pywhatkit as pw

import requests


speaker = win32com.client.Dispatch("SAPI.SpVoice")

def speak(audio):
    speaker.Speak(audio)

def wishme():
    hour = int(time.strftime("%H"))
    if 0 <= hour <= 11:
        speak("good morning")
    elif 12 <= hour <= 15:
        speak("good afternoon")
    elif 16 <= hour <= 20:
        speak("good evening")
    elif 21 <= hour <=23:
        speak("good night")
    speak("I am your AI assistant Jarvis sir. Please tell me how I may help you today.")

def get_location():
    """Get user's geolocation using IP address"""
    try:
        response = requests.get('https://ipapi.co/json/')
        data = response.json()
        lat = data.get('latitude')
        lon = data.get('longitude')
        return lat, lon
    except:
        speak("unable to fetch location")
        return 0, 0

def takecommand():
    recognizer = sr.Recognizer()
    fs = 44100
    duration = 6
    recognizer.energy_threshold=50
    recognizer.dynamic_energy_threshold=True
    
    

    print("listening...")
    
    
    audio_data = sd.rec(int(duration*fs), samplerate=fs, channels=1, dtype='int16')
    sd.wait()

    
    audio_float = audio_data.astype(np.float32)

    
    audio_float = audio_float / (np.max(np.abs(audio_float)) + 1e-8)

    
    audio_clean = (audio_float * 32767).astype(np.int16)

    
    audio_bytes = audio_clean.tobytes()

    
    audio = sr.AudioData(audio_bytes, fs, 2)

    try:
        print("recognizing...")
        text = recognizer.recognize_google(audio, language="en-in")
        print("you said:",text)
        return text.lower()
    except:
        print("Say again pls")
        return ""
    

wishme()
while True:

    query=takecommand().lower()
    if query =="":
        continue

    if "wikipedia" in query:

        speak("searching wikipedia...")
        query=query.replace("wikipedia","")  
        results=wikipedia.summary(query,sentences=5)
        speak(results)
        print(results)


    elif "open youtube" in query:
        speak("opening youtube") 
        webbrowser.open("youtube.com")


    elif "open google" in query:
        speak("opening google")
        webbrowser.open("google.com")


    elif "open insta" in query:
        speak("opening instagram")
        webbrowser.open("instagram.com")


    elif "open whatsapp" in query:
        webbrowser.open("whatsappweb.com")


    elif "mail" in query:
        speak("opening gmail")
        webbrowser.open("gmail.com")


    elif "shopping" in query:
        speak("opening flipkart")
        webbrowser.open("flipkart.com")


    elif "eat" in query:
        speak("lets order smthing to eat")
        webbrowser.open("swiggy.com")


    elif "time" in query: 
        speak(time.strftime("%H:%M:%S"))


    elif "play music" in query:
        speak("Please tell me the name of song you want to play?")
        song=takecommand().lower()
        speak(f"playing {song} on YT music")
        webbrowser.open("https://music.youtube.com/search?q="+song.replace("","%20"))


    elif "news" in query:
        speak("showing latest news")
        webbrowser.open("abpnews.in")



    elif "help" in query:
        speak("sending your live location on whatsapp to police")
        lat,lon=get_location()
        maps=f"https://www.google.com/maps?q={lat},{lon}"
        pw.sendwhatmsg_instantly("+919123713242",f"emergency!i need help my location {maps}",wait_time=10,tab_close=True)  



    elif "exit" in query:
        speak("thank u have a nice day")
        break






        
    










    
    


```
